## Dataframe engineering

**Methodology**

Schema:
Initially the following will be done on each dataframe
1. Check if all variables are unique and return a variable list
2. Check the variable types and memory usage. If variable is Pandas objects, verify if variable is purely non-numerical.
3. Determine the Shape of the dataframe
4. Determine the states of the non-numerical variables. If there are any inconsistencies it will be corrected. 
5. Check for duplicated rows. If any keep first occurence and drop the rest. Also return the new shape of the dataframe.
6. Check for missing values and conduct data imputation based on any available insights
7. Merge dataframes on available primary keys
8. Repeat steps 1-5


**Defining user defined functions**

```python
def df_variable_list(data):
    """
    This function checks if all the variables have unique names and 
    then returns a list of the dataframe variables.
    """
    if data.columns.is_unique:
        variable_list = data.columns.tolist()
        print("The dataframe variables are:\n", variable_list)
    else:
        print("There are variables with the similar names")


def df_object_vars(data):
    """
    This function checks if the variable type is pandas Object and
    examines if the values are only non-numerical in nature.
    It will highlight which variable(s) consisting of mixture of numerical and
    non-numerical values.
    """
    # Initialize a variables to inspect list
    vars_to_inpect = []

    # Loop though all variables
    for var in data.columns:
        # Check for pandas Object type
        if data[var].dtypes == "object":
            # Check for any numerical values
            if data[var].str.isnumeric().sum() != 0:
                vars_to_inpect.append(var)

    # Check if any variables of concern
    if len(vars_to_inpect) > 0:
        for variable in vars_to_inpect:
            print(
                f"The variable '{variable}' requires further exploration.")
    else:
        print("All pandas Object type variables are non-numerical")


def df_shape(data):
    """
    This function returns the shape of the dataframe.
    """
    shape = data.shape
    n_rows = shape[0]
    n_cols = shape[1]
    print(f"The dataframe  consists of '{n_rows}' rows and '{n_cols}' columns")


def df_duplicate_removal(data):
    """
    This function checks if there are any duplicated rows in the dataframe.
    If any, it displays the rows, keep the first occurence and drops the 
    duplicates.

    The new dataframe shape is calculated and returned
    """
    # Check if duplicated rows exist
    if any(data.duplicated() == True):
        print("The following is/are the duplicated row(s) in the dataframe")
        # Displays the duplicated row(s)
        display(data[data.duplicated() == True])
        # Drops the duplicates inplace while keeping the first occurence
        data.drop_duplicates(keep="first", inplace=True)

        # Check and returns the shape of the new dataframe
        new_df_shape = data.shape
        n_rows = new_df_shape[0]
        n_cols = new_df_shape[1]
        print(
            f"The new dataframe  consists of '{n_rows}' rows and '{n_cols}' columns")
    else:
        print("There is/are no duplicated row(s) in the dataframe")


def pandas_object_states(data):
    """
    This function checks if the variable type is pandas Object and
    displays the states and counts of each
    """
    # Loop through all variables
    for var in data.columns:
        # Check for pandas Object type
        if data[var].dtypes == "object":
            print('Unique values in', var, 'are :')
            print(data[var].value_counts().sort_index())
            print('--'*55)


def pandas_object_type_conv(data):
    """
    This function checks if the variable type is pandas Object and
    converts it to category type
    """
    # Loop through all variables
    for var in data.columns:
        # Check for pandas Object type
        if data[var].dtypes == "object":
            # Convert variable datatype to category
            data[var] = data[var].astype("category")

    # Confirm if there no variables with "object" type
    data.info()


def missing_val_chk(data):
    """
    This function to checks for missing values 
    and generates a summary.
    """
    if data.isnull().sum().any() == True:
        # Number of missing in each column
        missing_vals = pd.DataFrame(data.isnull().sum().sort_values(
            ascending=False)).rename(columns={0: '# missing'})

        # Create a percentage missing
        missing_vals['percent'] = ((missing_vals['# missing'] / len(data)) *
                                   100).round(decimals=3)

        # Remove rows with 0
        missing_vals = missing_vals[missing_vals['# missing'] != 0].dropna()

        # display missing value dataframe
        print("The missing values summary")
        display(missing_vals)
    else:
        print("There are NO missing values in the dataframe")
```

**Plotting functions**

```python
def pie_chart(data, figsize=(18, 16)):
    """
    This function prints a pie chart given a pandas Series.
    """

    label_pie = pd.Series(data.value_counts()).index.to_list()
    plt.pie(data.value_counts(),
            autopct='%1.1f%%',
            labels=label_pie,
            pctdistance=0.6,
            labeldistance=1.1,
            startangle=40,
            textprops={'fontsize': 14})
    plt.title("Distribution of {}".format(data.name),
              loc="center",
              fontsize=22)
    plt.legend(frameon=False, bbox_to_anchor=(0.85, 0.75))
    plt.show()
```

```python
# Team1 column played summary
team1_played_dist = matches_df.team1.value_counts()
# Team2 column played summary
team2_played_dist = matches_df.team2.value_counts()
# Combined team played
overall_team_played_dist = team1_played_dist + team2_played_dist
overall_team_played_dist.sort_values(ascending=False, inplace=True)
matches_played = overall_team_played_dist.to_frame().rename(
    {0: "matches_played"}, axis=1)

# Team win summary
matches_won = matches_df.winner.value_counts().to_frame().rename(
    {"winner": "matches_won"}, axis=1)

# Merging team played and team win summary
overall_match_summary = pd.merge(
    matches_played, matches_won, left_index=True, right_index=True)

# Adding "win_ratio" series
overall_match_summary["win_ratio"] = round(
    (overall_match_summary["matches_won"]/overall_match_summary["matches_played"])*100, 1)

# Sorting by "win_ratio"
overall_match_summary.sort_values(
    by=['win_ratio'], ascending=False, inplace=True)

display(overall_match_summary.style.highlight_max(color='lightgreen'))
```
